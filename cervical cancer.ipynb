```python
from sklearn.model_selection import train_test_split
import xgboost as xgb
```

```python
target=data['Biopsy'].astype('float32')
independent=data.drop(['Biopsy'],axis=1).astype('float32')
```

```python
x_train,x_test,y_train,y_test=train_test_split(independent,target,test_size=0.3)
```

```python
model=xgb.XGBClassifier(learning_rate=0.1,maximum_depth=50,n_estimators=100)
```

```python
model.fit(x_train,y_train)
```

```python
model.score(x_test,y_test)
```

```python
predicted=model.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix
accuracy=accuracy_score(y_test,predicted)
```

```python
accuracy
```

```python
cm=confusion_matrix(y_test,predicted)
```

```python
cm
```

```python
cm=pd.DataFrame(cm,columns=['predicted 0','predicted 1'],index=['actual 0','actual 1'])
```

```python
cm
```

```python
sns.heatmap(cm,annot=True,fmt='.0f',linewidths=0.1)
```
